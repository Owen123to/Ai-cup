```python
import numpy as np
import h5py
import random
import os
from IPython.display import clear_output
```

```python
def rotate90(y):
    row = y // 19
    col = y % 19
    rot_row = col
    rot_col = 18 - row
    return rot_row * 19 + rot_col

def mirror_hor(y):
    row = y // 19
    col = y % 19
    rot_row = row
    rot_col = 18 - col
    return rot_row * 19 + rot_col

def mirror_ver(y):
    row = y // 19
    col = y % 19
    rot_row = 18 - row
    rot_col = col
    return rot_row * 19 + rot_col
```
